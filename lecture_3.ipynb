```python
import pandas as pd
house = pd.read_csv("https://raw.githubusercontent.com/notice4/google-colab-study/refs/heads/main/housing.csv")
house.head()
```

```python
import pandas as pd
from  geopy.distance import geodesic

latt_center = house['latitude'].mean()
long_center = house['longitude'].mean()
center = (latt_center,long_center)
house['distance'] = house.apply(lambda x: geodesic((x['latitude'],x['longitude']),center).km,axis=1)

house.head()
```

```python
house.drop(['longitude','latitude'],axis=1,inplace=True)

house.head()
```

```python
house['Non_Bedrooms'] = house['total_rooms'] - house['total_bedrooms']
house.drop(['total_rooms'],axis=1,inplace=True)
house.head()
```

```python
import seaborn as sns
from sklearn.preprocessing import PowerTransformer

house.drop(['ocean_proximity'],axis=1,inplace=True)

for column in house.columns:
  if column!='median_house_value':
    pt = PowerTransformer()
    house[column] = pt.fit_transform(house[[column]])
house.head()
```
