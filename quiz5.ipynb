```python
import gradio as gr
from transformers import pipeline

model = pipeline("sentiment-analysis")

def pipeline(comment):
    if comment.strip() == "":
        return "ცარიერი ინფუთი ვერ შეფასდება... შეიყვანეთ ტექსტი"

    result = model(comment)[0]

    label = result["label"]
    score = result["score"] * 100

    if score > 90:
        score_message = "Score passed 90%"
    else:
        score_message = "Score have not passed 90%"

    return f"""
    Emotion: {label}
    Score: {score:.2f}%
    Result: {score_message}
    """

interface = gr.Interface(
    fn=pipeline,
    inputs=gr.Textbox(lines=5, placeholder="კომენტარი შეიყვანეთ აქ..."),
    outputs=gr.Textbox(lines=5, placeholder="ემოციური შეფასება მიიღეთ აქ"),
    title="Sentiment Analysis Program!",
    description="შეიყვანე კომენტარი და მიიღე სენტიმენტური ანალიზი!!!"
)

interface.launch()
```
