```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models

# 1. Define Paths
train_dir = 'medical/train'
test_dir = 'medical/test'

# 2. Initialize Data Generators
# We usually normalize pixel values to [0, 1] by dividing by 255
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,      # Optional: Data Augmentation
    horizontal_flip=True,   # Optional: Data Augmentation
    fill_mode='nearest'
)

# Note: We don't augment test data, only rescale it
test_datagen = ImageDataGenerator(rescale=1./255)

# 3. Load Images from Directory
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),  # Resize all images to 224x224
    batch_size=32,
    class_mode='categorical', # Use 'categorical' for 3+ classes
    shuffle=True
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=False             # Usually False for testing/evaluation
)

# Check the class mapping
print(f"Class Indices: {train_generator.class_indices}")
model = models.Sequential([
    # Convolutional layers to extract features
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # Flatten and Dense layers for classification
    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.Dropout(0.5),  # Helps prevent overfitting
    layers.Dense(3, activation='softmax')  # 3 classes
])

# 2. Compile the Model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=10, # Adjust based on your needs
    validation_data=test_generator,
    validation_steps=test_generator.samples // test_generator.batch_size,
    verbose=1
)
```
