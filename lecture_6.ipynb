### დენდროგრამის დიაგრამა

```python
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
X,_ = make_blobs(n_samples=3000, n_features=2, centers=[[2, 8], [7, 17], [11, 23]], cluster_std=1.2)
scores = AgglomerativeClustering(n_clusters=3).fit_predict(X)
print(silhouette_score(X, scores))
print(np.unique(scores))
```

```python
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
distance = linkage(X, method='ward')
dendrogram(distance)
```
